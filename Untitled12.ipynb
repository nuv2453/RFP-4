```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# Generate random cumulative rewards
num_time_steps = 10000
cumulative_rewards = np.cumsum(np.random.rand(num_time_steps))

# Normalize cumulative rewards to range from 0 to 1.0
cumulative_rewards /= cumulative_rewards.max()

# Increase the slope by multiplying y-values by 4
cumulative_rewards *= 6

# Interpolate the data points for a smoother curve
x = np.arange(num_time_steps)
y = cumulative_rewards
x_new = np.linspace(x.min(), x.max(), 300)  # 300 represents number of points to make the curve smooth
spl = make_interp_spline(x, y, k=3)  # Cubic spline interpolation
cumulative_rewards_smooth = spl(x_new)

# Plot the graph with curved line
plt.scatter(x, y, s=5)
plt.plot(x_new, cumulative_rewards_smooth, linestyle='-', alpha=0.7)
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Reward')
plt.title('Cumulative Reward over Time')
plt.xlim(0, num_time_steps)
plt.ylim(0, 6)  # Adjusted to match the multiplied range
plt.grid(True)
plt.show()
```
